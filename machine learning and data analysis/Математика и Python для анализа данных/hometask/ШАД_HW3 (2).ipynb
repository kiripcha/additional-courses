### Вычисление выборочных средних

```python
def get_sample_mean(data, N):
  np.random.shuffle(data)
  idx = np.random.randint(0,len(data)-N, 1000)
  sample_mean = []
  for id in idx:
    mean = np.mean(data[id:id+N])
    sample_mean.append(mean)
  return sample_mean
```

### Анализ ЦПД

```python
import scipy.stats as sps

y = np.random.lognormal( mean = 0, sigma = 1, size = 10000)
plt.figure(figsize = (10,10))
N = 1
grid = np.linspace(-3, 3, 1000) 
dat = get_sample_mean(y, N=N)
plt.subplot(3,1,1)
plt.plot(dat)
plt.grid()
plt.title("Выборочные средние для выборки %d"%N, fontsize=16)


plt.subplot(3,1,2)
plt.title("График плотности выборочных средних для выборки %d"%N, fontsize=16)
# plt.hist(dat,bins = 10,density=True)
mu, sigma = sps.norm.fit(y)
points = np.linspace(sps.norm.ppf(0.01,loc=mu,scale=sigma),
                 sps.norm.ppf(0.9999,loc=mu,scale=sigma),30)
pdf = sps.norm.pdf(points,loc=mu,scale=sigma)
h = plt.hist(dat,bins = 30,density=True)[0]
plt.plot(points, pdf, color='r')
plt.tight_layout()
plt.grid()

# Строим биплот для сравнения распределений
plt.subplot(3,1,3)
plt.title("Сравнение теоретического и фактического PDF для выборки %d"%N, fontsize=16)

x = (pdf - np.min(pdf))/(np.max(pdf)-np.min(pdf))
y_ = (h - np.min(h))/(np.max(h)-np.min(h))

plt.scatter(x, y_)
plt.plot([0,1],[0,1],color = 'y')
plt.xlabel("pdf-theoretical", fontsize=16)
plt.ylabel("pdf-sample mean", fontsize=16)

plt.tight_layout()
plt.grid()

print(sigma, np.mean(dat))
```

## Выводы

C увеличением N распределение становится больше похоже на нормальное (это видно из биплотов), при меньшем же N - конечное распределение больше похоже на исходное (можно взять крайний случай N = 1). Тем не менее, параметры nu и sigma получаются ближе к теоретическим, когда N минимально, поэтому такая апроксимация лучше.
